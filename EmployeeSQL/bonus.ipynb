```python
# Dependencies
import psycopg2
import numpy as np
import pandas as pd
import dbconfig as cfg
import pandas.io.sql as sqlio
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
# DB connection link
engine = create_engine(f"postgresql://{cfg.mysql['user']}:{cfg.mysql['passwd']}@{cfg.mysql['host']}:{cfg.mysql['port']}/{cfg.mysql['db']}")

# Connect to DB
conn = engine.connect()

# Extract salary data from DB
sql = "SELECT * FROM salaries ORDER BY salary ASC;"
df_sal = pd.read_sql(sql, conn)

# Extract avg salary by job title
sql = "SELECT titles.title, AVG(salaries.Salary) AS Avg_Salary FROM salaries "
sql = sql + "INNER JOIN titles ON titles.emp_no = salaries.emp_no GROUP BY titles.title ORDER BY AVG(salaries.Salary) ASC;"
df_avgsal = pd.read_sql(sql, conn)
                       
# Extract my salary
sql = "SELECT emp_no, first_name, last_name FROM employees WHERE emp_no = 499942;"
df_mysal = pd.read_sql(sql, conn)

conn.close()
```

```python
# Extract min and max salaries
min_sal = '${:0,.0f}'.format(df_sal["salary"].min())
max_sal = '${:0,.0f}'.format(df_sal["salary"].max())

print(f"Minimum Salary: {min_sal}")
print(f"Maximum Salary: {max_sal}")
```

```python
# Create histogram of salary ranges for all employees

fig=plt.figure(figsize=(12,8))
plt.hist(df_sal["salary"])
plt.xlabel("Salary", fontsize=12)
plt.ylabel("Employee Count by Salary",fontsize=12)
plt.ylim(0, 160000)
plt.title("Employee Salary Ranges", fontsize=15)
plt.show()
```

```python
# Create a bar chart of average salary by title
fig=plt.figure(figsize=(12,8))
plt.bar(df_avgsal["title"], df_avgsal["avg_salary"])
plt.ylim(0, 70000)
plt.title("Average Salary By Job Title", fontsize=15)
plt.xlabel("Job Title", fontsize=12)
plt.ylabel("Average Salary", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
